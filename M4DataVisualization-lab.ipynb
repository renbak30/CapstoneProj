## Demo: How to work with database

Connect to the database.


```python
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

Import pandas module.


```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name 
FROM sqlite_master 
WHERE type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master;
"""
pd.read_sql_query(QUERY,conn)
df = pd.read_sql_query(QUERY,conn)
df.head()


```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
pd.read_sql_query(QUERY,conn)
df_Age = pd.read_sql_query(QUERY,conn)
df_Age.plot(kind='box', figsize=(8, 6))
plt.title('Box Plot of Age')
plt.ylabel('Frequency')
plt.xlabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
pd.read_sql_query(QUERY,conn)
df_AgeWW = pd.read_sql_query(QUERY,conn)
df_AgeWW.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='darkblue')

plt.title('Age and Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')

plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
# the read_sql_query runs the sql query and returns the data as a dataframe

df = pd.read_sql_query(QUERY,conn)
df.head()
pd.read_sql_query(QUERY,conn)
df_BB = pd.read_sql_query(QUERY,conn)
sns.scatterplot(data=df_BB, x="CodeRevHrs", y="WorkWeekHrs", size="Age", legend=False, sizes=(5, 3000))

plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# SQL query to fetch all data from the table named 'DatabaseDesireNextYear'
QUERY = """
SELECT * 
FROM DatabaseDesireNextYear;
"""
pd.read_sql_query(QUERY,conn)
df = pd.read_sql_query(QUERY,conn)
# Execute the query and fetch data into a DataFrame

df = pd.read_sql_query(QUERY,conn)
df['DatabaseDesireNextYear'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.show()



```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
WHERE Age>=30 and Age<=35;
"""
pd.read_sql_query(QUERY,conn)
df = pd.read_sql_query(QUERY,conn)
df_Stack = df.groupby(['Age'])['WorkWeekHrs', 'CodeRevHrs'].median().plot.bar(stacked=True)


```

### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age>=45 and Age<=60;
"""
pd.read_sql_query(QUERY,conn)
df = pd.read_sql_query(QUERY,conn)
df_Line = df.groupby(['Age'])['ConvertedComp'].median().plot(kind='line')
plt.show()






```

### Bar Chart

Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch
FROM master;
"""
pd.read_sql_query(QUERY,conn)
df = pd.read_sql_query(QUERY,conn)
df['MainBranch'].value_counts().plot(x='MainBranch', y='val', kind='barh')

```
